```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
X_train.shape
```

SUPPORT VECTOR MACHINE

```python
classifier = svm.SVC(kernel="linear", gamma = "auto", C=2)
classifier.fit(X_train, y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predict))
print("the SVM model does not give a good value for the recall, and thus the F1 value is also weak. Overall it depends on how much cost can  be borne by the company to produce a bike that may not be sold. If the company can afford false positive result then we can go with the precision.")
```

```python
from sklearn.metrics import roc_curve, auc

svm_fpr, svm_tpr, threshold = roc_curve(y_test, y_predict)
auc_svm = auc(svm_fpr, svm_tpr)

plt.figure(figsize = (5,5), dpi=100)
plt.plot(svm_fpr, svm_tpr, linestyle = "-", label="svm (auc = %0.3f)" % auc_svm)
plt.xlabel("False-Positive Rate")
plt.ylabel("True-Positive Rate")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```
